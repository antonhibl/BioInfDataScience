```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import sklearn, sys, socket, requests
import altair as alt
import biom, qiime2
from qiime2 import Metadata 
from qiime2.plugins import feature_table, diversity, demux
from qiime2.plugins.demux.methods import emp_single
from qiime2.plugins.feature_table.visualizers import summarize as featsum
from qiime2.plugins.demux.visualizers import summarize
from qiime2.plugins.metadata.visualizers import tabulate
from qiime2.plugins.dada2.methods import denoise_single
from qiime2.plugins.diversity.pipelines import core_metrics
```

```python
my_artifact = qiime2.Artifact.load('./emp-single-end-sequences.qza')
metabars = qiime2.Metadata.load('sample-metadata.tsv')
html_metadata, = tabulate(metabars)
```

```python
barcode_data = metabars.get_column('barcode-sequence')
alphadiv_empsingles = emp_single(my_artifact, barcodes=barcode_data)
```

```python
summarize(alphadiv_empsingles.per_sample_sequences).visualization
```

```python
denoised_seq_data = denoise_single(alphadiv_empsingles.per_sample_sequences, trunc_len=92)
denoised_seq_data
```

```python
denoised_panda = denoised_seq_data.table.view(pd.DataFrame)
feat_sum = featsum(denoised_seq_data.table, sample_metadata=metabars)
feat_sum.visualization
```

```python
goku_data = core_metrics(table=denoised_seq_data.table, sampling_depth=1600, metadata=metabars)
```

```python
goku_panda = goku_data.rarefied_table.view(pd.DataFrame)
goku_panda
```

```python
shannon_series = goku_data.shannon_vector.view(pd.Series)
```

```python
otus_series = goku_data.observed_features_vector.view(pd.Series)
```

```python
evenness_series = goku_data.evenness_vector.view(pd.Series)
```

```python
series_table = pd.concat([shannon_series, otus_series, evenness_series], axis=1)
series_table
```

```python
metaframe = metabars.to_dataframe
metabars.to_dataframe()
```

```python
sigma_table = pd.concat([series_table, metabars.to_dataframe()], axis=1)
```
